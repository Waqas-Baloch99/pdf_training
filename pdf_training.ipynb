```python
from langchain.document_loaders import PyPDFLoader
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
import os
from dotenv import load_dotenv

load_dotenv()

google_api_key = os.getenv("GOOGLE_API_KEY")

# Load the PDF document
loader = PyPDFLoader("Wonka Chocolate Facility Rules.pdf")  # Update path if needed
documents = loader.load()

# Extract text from all pages
document_text = "\n".join([doc.page_content for doc in documents])

# Define the prompt template
template = """You are an AI assistant that answers questions based on the provided document.

Document:
{document}

Question: {question}
Answer:"""

prompt = PromptTemplate(
    input_variables=["document", "question"],
    template=template
)

llm = ChatGoogleGenerativeAI(model="gemini-pro")

qa_chain = LLMChain(llm=llm, prompt=prompt)

while True:
    query = input("\nAsk a question (type 'end' to exit): ").strip()
    if query.lower() == "end":
        break
    if not query:
        print("Please enter a valid question.")
        continue
    answer = qa_chain.run({"document": document_text, "question": query})
    print("\nAnswer:", answer)
```
